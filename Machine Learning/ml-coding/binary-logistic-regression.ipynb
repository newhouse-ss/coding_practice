## First from understand the algorithm from the formula
The task of logistic regression is: Binary classification, but don't be confused by its name and task, actually the output of logistic model is continuous([0, 1]), but at the final stage we will seperate the digits into 2 categories with a threshold, for example tres = 0.5.

Logistic regression = Linear model + Sigmoid Activation + binary cross entropy.

$z = Xw+b$ --> $a = \frac{1}{1+e^{-z}}$ --> $BCE = -\frac{1}{N}\sum_{i = 1}^N [yloga+(1-y)log(1-a)]$  (also named Log Loss)

$\frac{\partial{J}}{\partial{w}} = \frac{1}{N}X^T(A-Y)$, $\frac{\partial{L}}{\partial{w}} = X^T(a-y)$  
$\frac{\partial{J}}{\partial{b}} = \frac{1}{N}\sum_{i=1}^N(A-Y)$  
absolutely same as linear regression!

```python
import numpy as np

class Logestic_Regression:
    def __init__(self, learning_rate, iterations):
        self.lr = learning_rate
        self.itr = iterations

        self.weight = None
        self.bias = None
    
    def sigmoid(self, z):
        denominator = 1+np.exp(-z)
        return 1/denominator
    
    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weight = np.zeros(n_features)
        self.bias = 0

        y = y.reshape(-1)
        for i in range(self.itr):
            z = np.dot(X, self.weight)+self.bias
            y_prediction = self.sigmoid(z) # forward propagation

            dw = (1/n_samples)*np.dot(X.T, (y_prediction - y))
            db = (1/n_samples)*np.sum(y_prediction - y) # back propagation

            self.weight-=self.lr*dw
            self.bias-=self.lr*db

            cost = (-1/n_samples)*(np.dot(y, np.log(y_prediction))+np.dot((1-y), np.log(1-y_prediction)))

            if i%100 == 0:
                print(f"epoch{i} with cost {cost}")
    
    def prediction(self, X):
        z = np.dot(X, self.weight)+self.bias
        return self.sigmoid(z)
    

if __name__ == "__main__":
    m = 1000
    n = 2
    X = np.random.randn(m, n)

    w_true = np.array([[2.0], [3.0]])
    b_true = 1.5

    noise = np.random.randn(m, 1)
    Z = np.dot(X, w_true) + b_true + noise

    A = 1/(1+np.exp(-Z))

    Y = (A>0.5).astype(int)

    print(X.shape)
    print(Y.shape)

    model = Logestic_Regression(0.04, 1000)
    model.fit(X, Y)
    print("Result:")
    print("The real weight and bias is (2.0, 3.0) and 1.5")
    print(f"Model's prediction: {(model.weight)} and {model.bias}")

    pred = model.prediction(X)
    print(pred)
```
