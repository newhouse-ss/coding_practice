The model of linear regression is: $y = Xw+b$, we need to find the optimal $w$ and $b$,  
in order to do this, we need to use the Cost function MSE: J = ${1 \over N} \sum{(y^*-y)^2}$,   
compute gradient: $\frac{\partial J}{\partial w}$ and $\frac{\partial J}{\partial b}$,  
As for MSE, $dw = \frac{2}{N} \sum x_i(y^*-y) =  \frac{2}{N}X^T(y^*-y)$, $db = \frac{2}{N}\sum(y^*-y)$


minimal Cost function-->gradient decent-->gradient calculation

```python
import numpy as np

class Linear_Regression:
    def __init__(self, learning_rate, iterations):
        self.lr = learning_rate
        self.itr = iterations

        self.weight = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weight = np.zeros(n_features)
        self.bias = 0

        for i in range(self.itr):
            y_prediction = np.dot(X,self.weight)+self.bias

            dw = (1/n_samples)*np.dot(X.T, (y_prediction - y))
            db = (1/n_samples)*np.sum(y_prediction - y)

            self.weight -= self.lr*dw
            self.bias -= self.lr*db

            Cost = (1/n_samples)*np.sum((y_prediction-y)**2)

            if i%100 == 0:
                print(Cost)
    
    def prediction(self, X):
        return np.dot(X, self.weight)+self.bias

if __name__ == "__main__":
    X = 2 * np.random.rand(100, 1)
    y = 4 + 3 * X.flatten() + np.random.randn(100) * 0.2 # True w=3, b=4
    model = Linear_Regression(0.01, 1000)
    model.fit(X, y)
    print(model.bias, model.weight[0])
```
